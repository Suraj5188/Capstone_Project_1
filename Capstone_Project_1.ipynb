#**Data Cleansing Techniques To Get Rid Of Unwanted Data**

```python
# Now we will remove unwanted columns

df=df.drop(['url','address','phone','reviews_list','dish_liked','menu_item'],axis=1)
```

```python
df.head()
```

```python
# Now we will check info how many null values & dtype

df.info()
```

remove duplicates

```python
# Now we will remove duplicates

df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

Cleaning the rate column

```python
# Now we can see that in rate column '/' we don't want that

df['rate'].unique()
```

```python
# Here we have to repalce new and nan with another

def fun_repalce(n):
    if(n=='NEW' or n=='-'):
        return np.nan
    else:
        n = str(n).split('/')
        n = n[0]
        return float(n)
    
df['rate'] = df['rate'].apply(fun_repalce)
```

```python
df['rate'].head()
```

```python
df.head()
```

```python
df['rate']
```

```python
# we have replace '/' still we have NaN value that we have to replace 

# for that we replace that value average i.e mean in rate column

df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
# Now we will is their any null values or not

df['rate'].isnull().sum()
```

```python
# Now we are dropping null values

df.dropna(inplace=True)
```

Cleaning the approx_cost(for two people)

```python
# In approx_cost(for two people) their 'comma(,)' we want to remove that

def fun_repalce_comma(n):
  
  n=str(n) # str() function converts the specified value into a string

  if ',' in n:
    n=n.replace(',','')
    return float(n)
  else:
    return float(n)

df['approx_cost(for two people)']=df['approx_cost(for two people)'].apply(fun_repalce_comma)
```

```python
df.head()
```

```python
df.info()
```

```python
# We have to change the name of 'approx_cost(for two people)' because we want only one person cost

df.rename(columns={'approx_cost(for two people)':'Cost(Per_Person)'},inplace=True)
```

```python
# Now we have two person cost but we want only per person cost

df['Cost(Per_Person)'].div(2)
```

```python
df.head()
```

```python
# Now we have to clean 'rest_type' feature 

df['rest_type'].value_counts()
```

```python
# Now we can see that some values are highest number of restaurents

# But some are very less number of restaurents so we have to cluster them in 'other'

# To performing operation we store 'rest_type' in  variable 'restaurent_type'

restaurent_type=df['rest_type'].value_counts(ascending=False)
```

```python
# Their are we have compare the how many 'rest_type'  (max=19010) let's take min=1000 

# To store the resaturents less than 1000

rest_less_than=restaurent_type[restaurent_type < 1000]
```

```python
rest_less_than
```

```python
# Their are we have compare the how many 'rest_type'  (max=19010) let's take min=1000 

def fun2(n):
  if n in rest_less_than:
    return 'others'
  else:
    return n

df['rest_type']=df['rest_type'].apply(fun2)
```

```python
# Now we will check 

df['rest_type'].value_counts()
```

# **7. Statistics: Mean, Median, Mode.**<br>
Mean, median, and mode are three kinds of "averages". There are many "averages" in statistics, but these are, I think, the three most common, and are certainly the three you are most likely to encounter in your pre-statistics courses, if the topic comes up at all.

1. The **"mean"** is the "average" you're used to, where you add up all the numbers and then divide by the number of numbers.

2. The **"median"** is the "middle" value in the list of numbers. To find the median, your numbers have to be listed in numerical order from smallest to largest, so you may have to rewrite your list before you can find the median.There are two methods to calculate this value for odd and even number of values in list.

3. The **"mode"** is the value that occurs most often. If no number in the list is repeated, then there is no mode for the list.
